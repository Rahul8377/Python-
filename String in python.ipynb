# STRING 

• A string is a sequence of characters that can be a combination of letters, numbers, and special characters. It can be declared in python by using single quotes, double quotes, or even triple quotes. These quotes are not a part of a string, they define only starting and ending of the string.

• Strings are immutable, i.e., they cannot be changed. Each element of the string can be accessed using indexing or slicing operations.


• It is an immutable data type, meaning that once you have created a string, you cannot change it. Strings are used widely in many different applications, such as storing and manipulating text data, representing names, addresses, and other types of data that can be represented as text.


```python
str1 = 'Hello Python'  
print(str1)  
```

# 1.capitalize() 

Python capitalize() method converts first character of the string into uppercase without altering the whole string. It changes the first character only and skips rest of the string unchanged.


```python
str = "python is good language"

str1 = str.capitalize()
print(str1)


str2 = "rahul"
str3=str2.capitalize()
print(str3)
```

# 2.Count()

It returns the number of occurences of substring in the specified range. It takes three parameters, first is a substring, second a start index and third is last index of the range. Start and end both are optional whereas substring is required.



```python
str = "my name is rahul amberia and i am learning python "

str1=str.count('m')
print(str1)
str2=str.count('a')
print(str2)
```

# 3.find()

Python find() method finds substring in the whole string and returns index of the first match. It returns -1 if substring does not match.

```python
str = "my name is rahul amberia and i am learning python "

str2= str.find('and')
print(str2)

str3=str.find("learning")
print(str3)
```

# 4.format() Method

Python format() method is used to perform format operations on string. While formatting string a delimiter {} (braces) is used to replace it with the value. This delimeter either can contain index or positional argument.

#syntax

format(*args, **kwargs)  

*args : substring
**kwargs : start index a range

```python
str1="rahul"

str2="very"

str =" {} is a {} good learner.".format(str1,str2)
print(str)
```

# 5.index() Method
Python index() method is same as the find() method except it returns error on failure. This method returns index of first occurred substring and an error if there is no match found.

It throws an exception if string is not found. It works same as find() method.



```python
str = "my name is rahul amberia and i am learning python "

str1 = str.index("and")
print(str1)
```

# 6.isalnum()

It returns true if the characters in the string are alphanumeric i.e., alphabets or numbers and there is at least 1 character. Otherwise, it returns false.


Python isalnum() method checks whether the all characters of the string is alphanumeric or not. A character which is either a letter or a number is known as alphanumeric. It does not allow special chars even spaces.



```python
str =" rahul is going to be a data scientist"
str2=str.isalnum()
print(str2)
```

```python
str = "Welcome123" # True  
str3 = "Welcome 123"

str2=str.isalnum()
print(str2)
str4=str3.isalnum()
print(str4)
```

# 7. isalpha()

It returns true if all the characters are alphabets and there is at least one character, otherwise False.


Python isalpha() method returns true if all characters in the string are alphabetic. It returns False if the characters are not alphabetic. It returns either True or False.

```python
str = " my name is rahul"

str1 = str.isalpha()
print(str1)
```

```python
str = "python"  
if str.isalpha() == True:  
    print("String contains alphabets")  
else: print("String contains other chars too.")  
```

```python
str = "python language"  
if str.isalpha() == True:  
    print("String contains alphabets")  
else: print("String contains other chars too.")  
```

# Advantages of String in Python:

Python has built-in support for strings, which means that you do not need to import any additional libraries or modules to work with strings. This makes it easy to get started with strings and reduces the complexity of your code.
